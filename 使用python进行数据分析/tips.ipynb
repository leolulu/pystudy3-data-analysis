```python
import pandas as pd
import numpy as np

tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips.head()
```

```python
tips.pivot_table(['tip','tip_pct'],index=['day','time'],columns='sex',aggfunc='sum',margins=True)
```
