```python
from wordcloud import WordCloud
import jieba
```

```python
with open('./text.txt','r') as f:
    text = f.read()
```

```python
text_gen = jieba.cut(text)
text = ' '.join(text_gen)
```

```python
w = WordCloud(font_path=r'C:\Fonts\simhei.ttf',background_color='white',width=1920,height=1080,max_words=100)
```

```python
w.generate(text)
```

```python
w.to_image()
```

```python
w.to_file('wordcloud.png')
```
