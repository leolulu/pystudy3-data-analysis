```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('./datasets/sex-offenders.csv')
```

```python
data.head()
```

```python
sns.relplot(data=data,x='AGE',y='WEIGHT',col='GENDER',style='RACE',hue='RACE')
```

```python
ax = sns.swarmplot(x='RACE',y='AGE',data=data,hue="GENDER")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 20)
```

```python
ax = sns.catplot(x='RACE',y='AGE',data=data,hue="GENDER",kind='violin',split=True)
ax.set_xticklabels(rotation=45)
```

```python
ax = sns.barplot(data=data,x='RACE',y='AGE',hue='GENDER')
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,8))
sns.boxplot(x='RACE',y='AGE',data=data,ax=ax[0],hue='GENDER')
sns.scatterplot(x='AGE',y='WEIGHT',data=data,ax=ax[1],hue='RACE')
```

```python
sns.jointplot(x='AGE',y='WEIGHT',data=data,height=12,kind='hex')
```

```python
sns.pairplot(data=data,height=6,hue='GENDER')
```

```python
sns.FacetGrid(data,row='AGE',col='WEIGHT')
```
