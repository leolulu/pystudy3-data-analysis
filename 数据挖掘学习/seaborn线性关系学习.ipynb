```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips[:5]
```

```python
sns.catplot(data=tips,x='day',kind='count',col='size',col_wrap=3,palette='ch:.25',hue='time')
```

```python
sns.violinplot(data=tips,x='day',y='tip',hue='smoker',split=True)
```

```python
tips[:5]
```

```python
sns.regplot(data=tips,x='total_bill',y='tip',ci=99)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',row='sex',hue='smoker')
```

```python
sns.lmplot(data=tips,x='size',y='total_bill',x_jitter=0.1,size='tip')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',hue='smoker',size='tip',style='smoker',height=8)
```

```python
sns.kdeplot(tips.total_bill,tips.tip,n_levels=30)
```
