```python
import pandas as pd
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
```

```python
b = load_boston()
lr = LinearRegression()
std1 = StandardScaler()
std2 = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(b.data,b.target)
```

```python
x_train = std1.fit_transform(x_train)
x_test = std1.transform(x_test)

y_train = std2.fit_transform(y_train.reshape(-1,1))
y_test = std2.transform(y_test.reshape(-1,1))
```

```python
lr.fit(x_train,y_train)
```

```python
pd.DataFrame({
    'predict':list(std2.inverse_transform(lr.predict(x_test))),
    'y_test': list(std2.inverse_transform(y_test))
})
```
