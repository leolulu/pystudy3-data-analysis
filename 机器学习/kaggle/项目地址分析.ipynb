```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction import DictVectorizer
```

```python
rf = RandomForestClassifier()
gc = GridSearchCV(rf,
                  {
                      'n_estimators':[50,120,250,300,500],
                      'max_depth':[5,10,15,20,25,30]
                  },
                  cv=10
)
dv = DictVectorizer()
```

```python
pd.options.display.max_columns = 99
```

```python
objects_info = pd.read_csv('objects201808.csv')
objects_info = pd.concat(
[objects_info.loc[:,'ATT':'BTT'],objects_info['OBJECTTYPE']],
    axis=1
)
objects_info.dropna(inplace=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(objects_info.loc[:,'ATT':'BTT'],objects_info['OBJECTTYPE'])
```

```python
x_train = x_train.to_dict(orient='records')
x_test = x_test.to_dict(orient='records')
```

```python
x_train = dv.fit_transform(x_train)
x_test = dv.transform(x_test)
```

```python
y_train = y_train.map({'混合业务双向项目':0,
'纯数字双向项目':1,
'混合业务单向项目':2,
'撤销项目':3,
'纯模拟项目':4,
'纯数字单向项目':5,
'问题项目':6})
y_test = y_test.map({'混合业务双向项目':0,
'纯数字双向项目':1,
'混合业务单向项目':2,
'撤销项目':3,
'纯模拟项目':4,
'纯数字单向项目':5,
'问题项目':6})
```

```python
gc.fit(x_train,y_train)
```
