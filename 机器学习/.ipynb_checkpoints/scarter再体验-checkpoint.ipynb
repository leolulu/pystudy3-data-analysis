```python
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
import seaborn as sns
```

```python
b= load_boston()
```

```python
a = pd.concat(
    [pd.DataFrame(b.data,columns=b.feature_names),
    pd.DataFrame(b.target,columns=['target'])],
    axis=1
)
a.head()
```

```python
fig = plt.figure()
```

```python
plt.scatter(x=a['LSTAT'],y=a['target'],marker='x',s=1)
```

```python
sns.relplot(data=a, x='LSTAT',size='target',aspect=1.75,y='AGE',hue='CHAS',row='RAD',legend='full')
```
