```python
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
import pandas as pd
```

```python
i = load_iris()
```

```python
a= pd.concat(
    [pd.DataFrame(i.data,columns=['s1','s2','s3','s4']),
    pd.DataFrame(i.target,columns=['target'])],
    axis = 1
)

```

```python
fig = plt.figure()
```
