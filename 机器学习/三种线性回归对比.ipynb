```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor,Ridge
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
```

```python
b = load_boston()
lr = LinearRegression()
sgd = SGDRegressor(alpha=1)
r = Ridge(alpha=0.05)
std1 = StandardScaler()
std2 = StandardScaler()
fig = plt.figure(figsize=(8,8))
```

```python
x_train,x_test,y_train,y_test = train_test_split(b.data,b.target,random_state=13)
```

```python
x_train = std1.fit_transform(x_train)
x_test = std1.transform(x_test)

y_train = std2.fit_transform(y_train.reshape(-1,1))
y_test = std2.transform(y_test.reshape(-1,1))
```

```python
lr.fit(x_train,y_train)
sgd.fit(x_train,y_train)
r.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
sgd.score(x_test,y_test)
```

```python
r.score(x_test,y_test)
```

```python
plt.scatter(range(127),
            std2.inverse_transform(lr.predict(x_test)),
            marker='.',s=10,c='r')
plt.scatter(range(127),
            std2.inverse_transform(sgd.predict(x_test)),
            marker='.',s=10,c='g')
plt.scatter(range(127),
            std2.inverse_transform(r.predict(x_test)),
            marker='.',s=3,c='b')
```

```python
plt.plot(range(127),
            std2.inverse_transform(lr.predict(x_test)),
            c='r',linewidth=1)
plt.plot(range(127),
            std2.inverse_transform(sgd.predict(x_test)),
            c='g',linewidth=1)
plt.plot(range(127),
            std2.inverse_transform(r.predict(x_test)),
            c='b',linewidth=1)
```

```python
# 均方误差
mean_squared_error(std2.inverse_transform(y_test),std2.inverse_transform(lr.predict(x_test)))
```

```python
mean_squared_error(std2.inverse_transform(y_test),std2.inverse_transform(sgd.predict(x_test)))
```

```python
mean_squared_error(std2.inverse_transform(y_test),std2.inverse_transform(r.predict(x_test)))
```
